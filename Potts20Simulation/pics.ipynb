```python
import numpy as np, matplotlib.pyplot as plt
from collections import OrderedDict as od
from scipy.special import comb as C
acc_score = od()
```

```python
def get_DOS(N):
    DOS = od()
    Z = 0
    for k in range(0, 401, 2):
        DOS[-k] = C(N, k)
        Z += C(N, k)
    for k in range(0, 401, 2):
        DOS[-k] /= (2 * Z)
    return DOS
```

```python
plt.rcParams.update({'font.size': 16})
data = get_DOS(400)

plt.figure(figsize=(16, 9))
fig, ax = plt.subplots()

ax.plot([x for x in data], [g for x, g in data.items()])
#plt.fill_between([-200, 0], [0.04, 0.04], hatch='/', edgecolor='r', facecolor="none")
ax.set_ylabel("g(E)", rotation=0, labelpad=20, fontsize=18)
ax.yaxis.set_label_coords(-0.06, 1.03)
ax.set_xlabel("E", fontsize=18)
#ax.grid()
fig.savefig("./images/DOS_picture_2.png", bbox_inches='tight')
```
