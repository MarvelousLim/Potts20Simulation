```python
import numpy as np, matplotlib.pyplot as plt
from collections import OrderedDict as od
```

```python
data3 = od()
with open("./datasets/xorworv2_L40_R131072_run1e3.txt", "r") as f:
    for line in f:
        U = int(line.split()[1])
        d = od()
        for E, freq in zip(line.split()[2::2], line.split()[3::2]):
            d[int(E)] = int(freq)
        data3[U] = d
```

```python
data4 = od()
with open("./datasets/xorworv2_L40_R131072_run1e4.txt", "r") as f:
    for line in f:
        U = int(line.split()[1])
        d = od()
        for E, freq in zip(line.split()[2::2], line.split()[3::2]):
            d[int(E)] = int(freq)
        data4[U] = d
```

```python
plt.figure(figsize=(16, 9))

for U in [-2400, -2000, -1600, -1000, -600, 0]:
    d3 = data3[U]
    plt.plot([x for x in d3], [d3[x] for x in d3], label="U = {}".format(U))


#d4 = data4[U]
#plt.plot([x for x in d4], [d4[x] for x in d4])
plt.title("cooling, L=40")
plt.grid()
plt.legend()
plt.show()
```
