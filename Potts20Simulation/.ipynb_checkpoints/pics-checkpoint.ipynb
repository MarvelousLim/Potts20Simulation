```python
import numpy as np, matplotlib.pyplot as plt
from collections import OrderedDict as od
from scipy.special import comb as C
acc_score = od()
```

```python
def get_DOS(N):
    DOS = od()
    Z = 0
    for k in range(0, 401, 2):
        DOS[-k] = C(N, k)
        Z += C(N, k)
    for k in range(0, 401, 2):
        DOS[-k] /= (2 * Z)
    return DOS
```

```python
plt.figure(figsize=(16, 9))
data = get_DOS(400)
plt.plot([x for x in data], [g for x, g in data.items()])
#plt.fill_between([-200, 0], [0.04, 0.04], hatch='/', edgecolor='r', facecolor="none")
plt.xlabel("E")
plt.ylabel("g(E)")
plt.grid()
plt.savefig("./images/DOS_picture_2.png", bbox_inches='tight')
```

```python
plt.figure(figsize=(16, 9))
data = get_DOS(400)
plt.plot([x for x in data if x > -250 and x <= -200], [g for x, g in data.items() if x > -250 and x <= -200])
plt.fill_between([-202, -200], [data[-202], data[-200]], hatch='/', edgecolor='r', facecolor="none")
plt.xlabel("E")
plt.ylabel(r"Efficacit\'e", loc="top", rotation="horizontal")

#plt.ylabel("g(E)", rotation=0, loc=None)
plt.grid()
plt.savefig("./images/DOS_slice_picture.png", bbox_inches='tight')
```

```python
plt.figure(figsize=(16, 9))
data = get_DOS(400)
plt.plot([x for x in data], [np.log(g) for x, g in data.items()])
#plt.fill_between([-200, 0], [0.04, 0.04], hatch='/', edgecolor='r', facecolor="none")
plt.xlabel("E")
plt.ylabel("ln g(E)")
plt.grid()
plt.savefig("./images/DOS_picture_3.png", bbox_inches='tight')
```
