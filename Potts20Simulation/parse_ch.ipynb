```python
import numpy as np, matplotlib.pyplot as plt
```

```python
data = {}
for name in ["./datasets/xorwor_L40_R524288_run0ch.txt"]:
    with open(name) as f:
        for line in f:
            try:
                U = line.split(" ")[0]
                nline = line[len(U) + 1:] #-1 for \n
                U = int(U)
                if U not in data:
                    data[U] = {}
            except:
                break
            array = []
            for pair in nline.split(";"):
                try:
                    size, frequency = map(int, pair.split(","))
                    if size in data[U]:
                        data[U][size] += frequency
                    else:
                        data[U][size] = frequency
                except:
                    pass
```

```python
N = 40 * 40
U = -2496
sum([size * freq for size, freq in data[U].items()]) / N
```

```python
U = -2496
N = 40 * 40
e = U / N
plt.figure(figsize=(16, 9))
s = sum([size * freq for size, freq in data[U].items()])
plt.plot([size for size, freq in data[U].items()], [freq / s for size, freq in data[U].items()], 'o', 
         label="energy per spin={}; number of hists={}".format(e, int(s/N)))


U = -2768
e = U / N
array = data[U]
s = sum([size * freq for size, freq in data[U].items()])
plt.plot([size for size, freq in data[U].items()], [freq / s for size, freq in data[U].items()], 'o', 
         label="energy per spin={}; number of hists={}".format(e, int(s/N)))


U = -2944
e = U / N
array = data[U]
s = sum([size * freq for size, freq in data[U].items()])
plt.plot([size for size, freq in data[U].items()], [freq / s for size, freq in data[U].items()], 'o', 
         label="energy per spin={}; number of hists={}".format(e, int(s/N)))

plt.plot([N, N], [0, 10**0], c='r')
plt.title("L={}, R={}".format(40, 32768))
plt.yscale("log")
plt.xscale("log")
plt.ylabel("frequency")
plt.xlabel("cluster size")
plt.legend()
plt.grid()
plt.show()
```
